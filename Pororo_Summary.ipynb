```
from pororo import Pororo
import pandas as pd
```

```
Pororo.available_tasks()
```

```
summary = Pororo(task="summary", lang="kr")
```

```
df = pd.read_csv('./data/summarize_2.csv')
#df = pd.read_csv('./data/tmp_res.csv')

df.head()
```

```
i = 7299

while i != 8000:
    res = summary(df['lyrics'][i])
    df['lyrics'][i] = res
    print("{}번째 요약: {}" .format(i, res))
    i += 1
```

```
df.to_csv('./data/summarize_2.csv', index = False, encoding='utf-8-sig')
```
