## 예제 - load_digits(손글씨 데이터)

```
import umap
import umap.plot
from sklearn.datasets import load_digits

digits = load_digits()

mapper = umap.UMAP().fit(digits.data)
umap.plot.points(mapper, labels=digits.target)
```

## 가사 그래프 - 멀티

- n_neighbors : 인접 점의 수
- min_dist: 값이 클수록 점이 더 고르게 분포, 값이 작을 수록 로컬 구조에 대해 더 정확하게 최적화
- metric: 입력 공간의 거리 측정 

```
import umap
import umap.plot

digits = mul

mapper = umap.UMAP().fit(digits.data)
umap.plot.points(mapper)
```

```
mapper = umap.UMAP(n_neighbors=4,
                   min_dist=0.01).fit(digits.data)
umap.plot.points(mapper)
```

```
mapper = umap.UMAP(n_neighbors=10,
                   min_dist=0.001).fit(digits.data)
umap.plot.points(mapper)
```

```
mapper = umap.UMAP(n_neighbors=10,
                   min_dist=0.5).fit(digits.data)
umap.plot.points(mapper)
```
