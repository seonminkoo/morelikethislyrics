## 유클리드 거리

----
값이 작을수록 유사함

```
def dist(x,y):   
    return np.sqrt(np.sum((x-y)**2))
```

```
for i, c in enumerate(cosvec_ko[:1000]):
  print("{}번째 노래와의 유사도 : {}" .format(i, dist(cosvec_ko[0], c)))
```
